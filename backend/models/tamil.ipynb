```python
from pypdf import PdfReader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
import uuid

# Load embedding model
embed_model = SentenceTransformer("intfloat/multilingual-e5-base")

def embed_texts(texts):
    texts = ["passage: " + t for t in texts]
    return embed_model.encode(texts, normalize_embeddings=True)

splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)
```

```python
with open(TAMIL_TXT, "r", encoding="utf-8") as f:
    tamil_text = f.read()

tamil_chunks = splitter.split_text(tamil_text)
```

```python
import chromadb
import uuid

# 1. Initialize Chroma
client = chromadb.Client(Settings(persist_directory="./chroma_db"))

# 2. Check the model dimension (sanity check)
test_embed = embed_model.encode(["test"])
current_dim = test_embed.shape[1]
print(f"рҹ“Ў Your model's actual dimension is: {current_dim}")

# 3. Force Delete and Recreate
try:
    client.delete_collection(name="science_rag_final")
except:
    pass

# Create a fresh collection
collection = client.create_collection(name="science_rag_final")

# 4. Corrected Add Function
def add_chunks_to_db(chunks, lang):
    # e5 models need the "passage: " prefix for indexing
    embeddings = embed_texts(chunks)

    # Convert numpy to list for ChromaDB compatibility
    embed_list = embeddings.tolist()

    ids = [str(uuid.uuid4()) for _ in chunks]
    metadatas = [{"language": lang} for _ in chunks]

    collection.add(
        ids=ids,
        documents=chunks,
        embeddings=embed_list,
        metadatas=metadatas
    )
    print(f"вң… Added {len(chunks)} {lang} chunks.")

# 5. Run the indexing
add_chunks_to_db(tamil_chunks, "tamil")
```

```python
from langdetect import detect

def retrieve_chunks(query):

    try:
        lang_code = detect(query)
        target_lang = "tamil" if lang_code == "ta" else "english"
    except:
        target_lang = "english"

    # E5 query format
    query_emb = embed_model.encode(
        ["query: " + query],
        normalize_embeddings=True
    ).tolist()

    results = collection.query(
        query_embeddings=query_emb,
        n_results=5,
        where={"language": target_lang}
    )

    documents = results["documents"][0]

    # рҹ”Һ Print retrieved chunks
    print("\n--- Retrieved Chunks ---\n")
    for i, doc in enumerate(documents):
        print(f"Chunk {i+1}:")
        print(doc)
        print("-" * 50)

    context = "\n".join(documents[:3])  # limit context

    return context
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

model_name = "abhinand/tamil-llama-7b-instruct-v0.1"

tokenizer = AutoTokenizer.from_pretrained(model_name)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4"
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto"
)

model.eval()
```

```python
def generate_answer(query):

    context = retrieve_chunks(query)

    prompt = f"""
а®ЁаҜҖа®ҷаҜҚа®•а®іаҜҚ а®’а®°аҜҒ а®…а®ұа®ҝа®өа®ҝа®Ҝа®ІаҜҚ а®Ҷа®ҡа®ҝа®°а®ҝа®Ҝа®°аҜҚ.

а®•аҜҖа®ҙаҜҮ а®үа®іаҜҚа®і а®Өа®•а®өа®ІаҜҲ а®Әа®Ҝа®©аҜҚа®Әа®ҹаҜҒа®ӨаҜҚа®Өа®ҝ а®•аҜҮа®іаҜҚа®өа®ҝа®•аҜҚа®•аҜҒ а®Әа®Өа®ҝа®ІаҜҚ а®…а®іа®ҝа®•аҜҚа®•а®өаҜҒа®®аҜҚ.

а®Өа®•а®өа®ІаҜҚ:
{context}

а®•аҜҮа®іаҜҚа®өа®ҝ:
{query}

а®Өа®®а®ҝа®ҙа®ҝа®ІаҜҚ а®ӨаҜҶа®іа®ҝа®өа®ҫа®• а®Әа®Өа®ҝа®ІаҜҚ а®…а®іа®ҝа®•аҜҚа®•а®өаҜҒа®®аҜҚ.
"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=200,
            temperature=0.7,
            do_sample=True
        )

    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return answer
```

```python
def clean_context(text):
    banned = [
        "а®Үа®ӨаҜҲ а®үа®ҷаҜҚа®•а®іаҜҚ а®ЁаҜҶа®ҹаҜҚа®өаҜҠа®°аҜҚа®•аҜҚ",
        "а®ҡаҜҮа®®а®ҝа®•аҜҚа®•а®өаҜҒа®®аҜҚ",
        "RIGHT COLUMN",
        "LEFT COLUMN"
    ]

    for b in banned:
        text = text.replace(b, "")

    return text
```

```python
query = "а®•а®ҫа®іа®ҫа®©аҜҚ а®Әа®ҫа®ұаҜҲ а®Һа®©аҜҚа®ұа®ҫа®ІаҜҚ а®Һа®©аҜҚа®©?"
answer = generate_answer(query)
print(answer)
```
